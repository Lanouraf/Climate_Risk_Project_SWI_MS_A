## Variogram
### Variogram Cloud

We begin with the **variogram cloud**, which helps us visualize the intrinsic variability of the dataset. Each point represents a pairwise semivariance computed between two observation points, shown before any binning or smoothing is applied.

```python
# Data: coords and values from your GeoDataFrame or sample DataFrame
coords = geodata_[["x", "y"]].values
values = geodata_["z"].values

# Compute pairwise distances (Euclidean distance by default)
coords_dist = pdist(coords)  # condensed distance matrix
# Compute pairwise squared differences (semi-variances)
values_diff = pdist(values[:, None])  # value differences
gamma = 0.5 * (values_diff ** 2) # theoretical gamma of the variogram

# Plot cloud
plt.figure(figsize=(8, 6))
plt.scatter(coords_dist, gamma, alpha=0.3, s=10)
plt.xlabel("Distance")
plt.ylabel("Semivariance")
plt.title("Variogram Cloud (Pointwise)")
plt.grid(True)
plt.show()
```

### Empirical Variogram
The class `Variogram` from the `scikit-gstat` library computes the empirical variogram from spatial data.
Its different parameters allow you to control the number of lag classes, their width, and the maximum distance used in the calculation, similar to the options of `variog` in `R`.

The method `.plot()` displays the resulting variogram.

Try adjusting the following parameters to explore how these affect the shape of the variogram and to find a configuration that best captures the spatial structure of your data
- `maxlag` (maximum distance — lag — between point pairs that are included when computing the empirical variogram),
- `bin_func` (binning function used to find lag class edges),
- `n_lags` (number of lag classes).

```python
V1 = Variogram(coords, values, normalize=False)
V1.plot(show=False);
```

```python
# Empirical variogram (default binning)
V = Variogram(
    coords,
    values,
    # normalize=False,
    # estimator='matheron',
    model='spherical',
    # dist_func='euclidean',
    fit_method='lm',  # controls how the theoretical variogram model is fitted to the empirical variogram points
    bin_func='even',   # equal-width bins
    # n_lags=10,  # default=10
    use_nugget=False  # default=False
)
```

```python
# Plot of just the empirical variogram
plt.figure(figsize=(6, 4))
plt.scatter(V.bins, V.experimental, marker='o', linestyle='-', label='Empirical variogram')
plt.xlabel("Distance (lag)")
plt.ylabel("Semivariance")
plt.title("Empirical Variogram (no model fit)")
plt.grid(True)
plt.legend()
plt.show()
```

We observe that the semivariance decreases at large distances.
This is not necessarily due to a true spatial correlation pattern,
but rather an _artifact_ caused by having very few point pairs at long distances.
As distance increases, the number of available pairs decreases rapidly,
making the semivariance estimate at those lags unstable and noisy.

For this reason, it is common practice to limit the variogram to a certain cutoff distance,
such as the `mean` or `median` of all pairwise distances.
This ensures that the empirical variogram is based on a sufficient number of pairs per bin
and avoids overinterpreting unreliable long-range behavior.

```python
plt.figure()
V.plot()
plt.title("Empirical Variogram")
plt.show()
```

```python
fixed_maxlag = 80
fixed_n_lags = 20
```

```python
# V.maxlag='median'
V.maxlag=fixed_maxlag
V.n_lags=fixed_n_lags
# V.bin_edges=np.arange(0, V.maxlag, 10)  # seq(0,80,10) equivalent

plt.figure()
V.plot()
plt.title("Empirical Variogram")
plt.show()
```

### Fitted Variogram
The `Variogram` class fits a theoretical variogram model to the empirical variogram.
Once computed, the model can be plotted together with the empirical points to visualize how well it reproduces the observed spatial variability.

Different options allow you to choose the model type (exponential, gaussian, spherical, …), the presence or absence of a nugget effect, and the initial parameter values used in the fitting algorithm.

By comparing the fitted curves obtained with different models and nugget settings, we can evaluate which configuration best represents the spatial structure of the data.

```python
# Fit one model to preview summary
V_fitted = copy.deepcopy(V)
V_fitted.model = 'exponential'
print("Fitted model parameters:", V_fitted.parameters)
```

```python
# Summary of the parameters' estimation
def variogram_summary(V):
    """
    Print a structured summary of a fitted Variogram object (scikit-gstat)
    """

    print("Variogram Model Summary\n" + "-" * 35)

    # Model name
    print("Used model:")
    print(f"  '{V.model.__name__}'\n")

    # Estimation method (approximation)
    print("Fit method:")
    print(f"  '{V.fit_method}'\n")

    # Spatial component (sill & range)
    print("Spatial components:")
    print(f"  sill (sigmasq) : {V.parameters[0]:.6f}")
    # partial sill (the variance explained by spatial structure) \neq  total sill (V.sill = V.nugget + V.parameters[0])
    print(f"  range (phi)    : {V.parameters[1]:.6f}\n")
    # range parameter (the distance at which spatial correlation becomes negligible)

    # Additional component (Matern parameter)
    if V.model.__name__ == 'matern':
        kappa = V.model.kwargs.get('kappa', 'not set')
        print("Extra spatial component:")
        print(f"  kappa   : {kappa}\n")

    # Nugget effect
    print("Estimated nugget:")
    print(f"  tausq   : {V.parameters[2]:.8f}\n")

    # Practical range (effective range)
    # print("Effective range:")
    # print(f"  {V.effective_range:.5f}\n")
    # distance at which the model reaches 95% of the sill,
    # when spatial dependence becomes negligible, independent of the model specification

    # Sum of squares (not available in scikit-gstat)
    # Empirical and model values
    empirical = V.experimental
    lags = V.bins
    nugget = V.parameters[2]
    
    # Predicted semivariances from the fitted model
    predicted = np.array([V.model(h, *V.parameters) + nugget for h in lags])
    
    # Optional: use weights (number of pairs per bin)
    weights = V.bin_count  # or np.ones_like(empirical) for unweighted SSR
    
    # Compute sum of squared residuals (weighted)
    ssr = np.sum(weights * (empirical - predicted) ** 2)
    print(f"Sum of squared residuals (SSR): {ssr:.4f}")
```

```python
variogram_summary(V)
```

```python
# ========== Fit multiple models with varying nugget & type ==========

# Note: scikit-gstat does not let you "fix" the nugget directly,
# but we can manually set parameters and skip fitting if needed.

# Model 1: exponential, estimated nugget
v1 = copy.deepcopy(V)
v1.model = 'exponential'
# v1.use_nugget=True
```

```python
# Model 2: exponential, nugget = 0.25

# Fix the nugget
fixed_nugget = 0.25

# Define model function (e.g., spherical) with nugget added
def variogram_model(h, sill, rng):
    return models.exponential(h, sill, rng) + fixed_nugget

# Fit sill and range to empirical data
popt, _ = curve_fit(
    variogram_model,
    V.bins,
    V.experimental,
    p0=[100., 2.]   # initial guesses for [sill, range]
)

# Inject results into a new Variogram object
v2 = Variogram(coords, values,
               model='exponential',
               fit_method='manual',
               fit_sill=popt[1],
               fit_range=popt[0],
               fit_nugget=fixed_nugget,
               maxlag=fixed_maxlag,
               n_lags=fixed_n_lags
              )
```

```python
# Model 3: gaussian, nugget = 0
v3 = copy.deepcopy(V)
v3.model = 'gaussian'
```

```python
# Model 4: gaussian, nugget = 0.25

# Define model function as gaussian with nugget added
def variogram_model(h, sill, rng):
    return models.gaussian(h, sill, rng) + fixed_nugget

# Fit sill and range to empirical data
popt, _ = curve_fit(
    variogram_model,
    V.bins,
    V.experimental,
    p0=[100., 2.]  # initial guesses for [sill, range]
)

# Inject results into a new Variogram object
v4 = Variogram(coords, values,
               model='gaussian',
               fit_method='manual',
               fit_sill=popt[1],
               fit_range=popt[0],
               fit_nugget=fixed_nugget,
               maxlag=fixed_maxlag,
               n_lags=fixed_n_lags
              )

# v4.plot()
```

```python
# Model 5: spherical, nugget = 0
v5 = copy.deepcopy(V)
v5.model ='spherical'
```

```python
# Model 6: spherical, nugget = 0.25

# Define model function as spherical with nugget added
def variogram_model(h, sill, rng):
    return models.spherical(h, sill, rng) + fixed_nugget

# Fit sill and range to empirical data
popt, _ = curve_fit(
    variogram_model,
    V.bins,
    V.experimental,
    p0=[100., 2.]  # initial guesses for [sill, range]
)

# Inject results into a new Variogram object
v6 = Variogram(coords, values,
               model='spherical',
               fit_method='manual',
               fit_sill=popt[1],
               fit_range=popt[0],
               fit_nugget=fixed_nugget,
               maxlag=fixed_maxlag,
               n_lags=fixed_n_lags
              )
```

```python
# ========== Plot all fitted models ==========

lags = np.concatenate(([0.],V.bins))  # use the same lags

# Define function to plot a model
def plot_model(vario, label, color, nugget):
    gamma_model = [vario.model(h, *vario.parameters) + nugget for h in lags]
    plt.plot(lags, gamma_model, label=label, color=color)

variograms_list = [v1,  v2,  v3,  v4,  v5,  v6]
label_list = ['spherical', 'exponential', 'gaussian', 'gaussian']
label_nugget_list = ['nugget 0', 'nugget 0.25']
nugget_list = [ 0., .25]
color_list = ['black', 'red', 'green', 'blue', 'purple', 'orange']

plt.figure(figsize=(10, 8))

for i in range(len(variograms_list)):
    label_ = label_list[i%3]
    label_nugget_ = label_nugget_list[i%2]
    nugget_ = nugget_list[i%2]
    plot_model(variograms_list[i],
         label_+" with "+label_nugget_,
         color_list[i],
         nugget=nugget_
        )

plt.scatter(v6.bins, v6.experimental, marker='o', linestyle='-', label='Empirical variogram')

plt.title("Different models")

plt.legend(loc='lower right', fontsize=9)
plt.grid(True)
plt.show()
```

## Kriging

The `OrdinaryKriging` class from the `pykrige` library performs kriging over a spatial grid based on the selected variogram model.
The method `.execute()` computes the **kriged surface** (predicted values) as well as the **kriging variance** at each grid point.

The grid of prediction locations is created using a mesh over the desired spatial domain. The kriging type (e.g., ordinary, universal) and model settings (e.g., variogram shape) are specified during the initialization of the `OrdinaryKriging` object.

The resulting kriged map can be visualized using `matplotlib.pyplot.imshow()`, which displays both the **spatial prediction** and its **uncertainty**.

From the different variogram models fitted earlier, one can now generate and compare the resulting kriging maps and associated variance maps.
This allows for discussion of the spatial structure captured by each model, and the sensitivity of the kriging results to model choice.

```python
def interpolate(V, ax):
    # Create prediction grid
    xx, yy = np.mgrid[0:100:100j, 0:100:100j]

    # Ordinary Kriging from scikit-gstat
    ok = OrdinaryKriging(V, min_points=5, max_points=15, mode='exact')
    field = ok.transform(xx.flatten(), yy.flatten()).reshape(xx.shape)

    # Show interpolated field
    art = ax.matshow(field, origin='lower', cmap='plasma', vmin=V.values.min(), vmax=V.values.max())

    # Add original data points on top
    ax.scatter(geodata_['x'], geodata_['y'], alpha=0.7, color='black', s=20, label='Observed data')

    # Plot config
    ax.set_title('%s model' % V.model.__name__)
    plt.colorbar(art, ax=ax)
    return field
```

```python
variograms_list_new = [v1, v3, v5, v2, v4, v6]
d_fields = []
fig, _a = plt.subplots(2,3, figsize=(18, 12), sharex=True, sharey=True)
axes = _a.flatten()
for i in range(len(variograms_list)):
    d_fields.append(interpolate(variograms_list_new[i], axes[i]))
```

## Threshold Exceedance Mapping via Kriging Simulation

We simulate multiple realizations of the spatial field using ordinary kriging with added stochastic variability. This allows us to estimate the **probability that the field exceeds a given threshold** (e.g., $Z > 2$) at each location on a prediction grid.

The steps include:
- Generating a prediction grid over the spatial domain.
- Performing multiple kriging simulations.
- Computing, for each grid point, the **empirical probability** that the simulated value exceeds the threshold.
- Visualizing both a simulated field (e.g., 75th percentile) and the **probability of exceedance** map.

This approach helps assess the **uncertainty and risk** of high values in spatial models, a useful tool for decision-making under uncertainty in fields like environmental science, insurance, and risk management.

```python
xg = np.arange(0, 100+2, 2)
yg = np.arange(0, 100+2, 2)
xx, yy = np.meshgrid(xg, yg)
grid_points = np.column_stack([xx.ravel(), yy.ravel()])
```

```python
n_sim = 100  # Number of simulated fields
threshold = 2

simulations = []

for _ in range(n_sim):
    ok = OrdinaryKriging(V, min_points=5, max_points=15, mode='exact')
    sim = ok.transform(xx.ravel(), yy.ravel())
    
    # Add Gaussian noise consistent with kriging variance
    # NOTE: This is a crude approximation of conditional simulation
    noise = np.random.normal(0, 1, size=sim.shape)
    simulations.append(sim + noise)

simulations = np.stack(simulations)  # shape: (n_sim, n_points)
```

```python
# Probability of exceeding threshold at each location
proba = (simulations > threshold).mean(axis=0)

# Reshape for plotting
simu_percentile = np.percentile(simulations, 75, axis=0).reshape(xx.shape)
proba_map = proba.reshape(xx.shape)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(simu_percentile, extent=[0, 100, 0, 100], origin='lower',
           cmap='Spectral_r', vmin=-4, vmax=4)
plt.colorbar(label="Simulated Z (75th percentile)")
plt.title("Conditional Simulation")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(geodata_['x'], geodata_['y'], color='black', s=10)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(1 - proba_map, extent=[0, 100, 0, 100], origin='lower',
           cmap='Spectral', vmin=0, vmax=1)
plt.colorbar(label="P(Z ≤ 2)")
plt.title("Probability of Not Exceeding Threshold")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(geodata_['x'], geodata_['y'], color='black', s=10)
plt.show()
```
